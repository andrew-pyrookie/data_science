```python
# modules we'll use
import pandas as pd
import numpy as np
import seaborn as sns
import datetime

# reading our data 
landslides = pd.read_csv('data/GLC03122015.csv')

# set seed for reproducibility
np.random.seed(0)
```

```python
landslides.head()
```

```python
# will be working with the date column

print(landslides.iloc[0]['date_'])
landslides['date_'].dtype
```

```python

def change_format(dates):
    dates = pd.to_datetime(dates)
    dates = dates.dt.strftime('%m/%d/%Y')
    return dates

landslides['date_'] = change_format(landslides['date_'])
print(landslides['date_'].head())
```

```python
# create column :

landslides['date_parsed'] = pd.to_datetime(landslides['date_'])
landslides['date_parsed'].head()
```
